```python
import pandas as pd

S3_FILE_PATH = "s3://public-health-dashboard/jhu_covid19/"

TESTING_URL = f"{S3_FILE_PATH}county-city-testing.parquet"
```

```python
df = pd.read_parquet(TESTING_URL)

df["date2"] = pd.to_datetime(df.date)

keep = ["date2", "County_Performed", "City_Site_Performed", "County_Site_Performed"]
df = df.dropna(subset = ["City_Site_Performed", "County_Site_Performed"])[keep]

df["Sites_Total"] = df.City_Site_Performed + df.County_Site_Performed
df["correct"] = df.apply(lambda row: True if row.Sites_Total <= row.County_Performed else False, axis=1)
```

```python
df[df.date2 < '7/1/20'].correct.value_counts()
```
